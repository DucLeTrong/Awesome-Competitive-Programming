```
from collections import defaultdict

'''
NOTE:
- I use "BFS-shortest-path" to solve this. Time complexity is O(V+E).
- If you use "Dijkstra", you can't pass all test cases. Time complexity is O(ELogV)).
'''

class Graph: 
  # Constructor 
  def __init__(self): 
    # Default dictionary to store graph 
    self.graph = defaultdict(list) 
  
  # Function to add an edge to graph 
  def addEdge(self,u,v): 
    # In case of disconnected graph
    if v == None:
      self.graph[u]
    # In case of connected graph
    else:
      self.graph[u].append(v) 

# Function to find shortest path based on BFS
# Time complexity = O(V+E)
def BFS_SP(graph, start, goal): 
  explored = [] 
      
  # Queue for traversing the graph in the BFS 
  queue = [[start]] 
      
  # Loop to traverse the graph with the help of the queue 
  while queue: 
    path = queue.pop(0) 
    node = path[-1] 
          
    # Condition to check if the current node is not visited 
    if node not in explored: 
      neighbours = graph[node] 
              
      # Loop to iterate over the neighbours of the node 
      for neighbour in neighbours: 
        new_path = list(path) 
        new_path.append(neighbour) 
        queue.append(new_path) 
                  
        # Condition to check if the neighbour node is the goal 
        if neighbour == goal: 
          return new_path

      explored.append(node)

# Main function
def countLuck(matrix, k):
  # Decompose each character in each string into
  # a single element
  mat = [] # new matrix
  for x in matrix:
    mat.append(list(x))
  
  # Set some initial variables
  NodeNumber = 0
  row = len(matrix)
  col = len(matrix[0]) 
  start = None # Starting node
  end = None # Destination node
  g = Graph() # Create a graph

  for i in range(0,row):
    for j in range(0,col):
      if mat[i][j] == 'X': # If the node is blocked
        g.addEdge(NodeNumber,None)

      elif mat[i][j] != 'X':
        # Find node number of the starting point
        if mat[i][j] == 'M':
          start = NodeNumber

        # Find node number of destination point
        elif mat[i][j] == '*':
          end = NodeNumber

        # Check if the node can be connected to others
        # based on 4 directions
        if i-1 >= 0:
          if mat[i-1][j] != 'X':
            g.addEdge(NodeNumber,NodeNumber-col)

        if i+1 <= row-1:
          if mat[i+1][j] != 'X':
            g.addEdge(NodeNumber,NodeNumber+col)

        if j-1 >= 0:
          if mat[i][j-1] != 'X':
            g.addEdge(NodeNumber,NodeNumber-1)

        if j+1 <= col-1:
          if mat[i][j+1] != 'X':
            g.addEdge(NodeNumber,NodeNumber+1)

      # Step to next node
      NodeNumber += 1

  # Find shortest path from 'M' to '*'
  path = BFS_SP(g.graph,start,end)
  
  # Walk on the path and check how many waves needed
  count = 0
  if len(g.graph[path[0]]) >= 2:
    count += 1
  for node in path[1:len(path)-1]:
    if len(g.graph[node]) >= 3:
      count += 1

  # Final conclusion
  if k == count:
    return 'Impressed'
  else:
    return 'Oops!'

if __name__ == '__main__':
  t = int(input())

  for t_itr in range(t):
    nm = input().split()

    n = int(nm[0])
    m = int(nm[1])

    matrix = []
    for _ in range(n):
      matrix_item = input()
      matrix.append(matrix_item)

    k = int(input())

    result = countLuck(matrix, k)
    print(result)
```
