```
# Binary Search Tree Algorithm to Find a Number of a Sorted List
# Time complexity = O(logn)
from bisect import bisect_left

def BST_Find(sortedList,n):
  idx = bisect_left(sortedList,n)
  if idx < len(sortedList) and sortedList[idx] == n:
    return True
  else:
    return False

# Binary Search Tree Algorithm to Find the Index of a Number of a Sorted List
# Time complexity = O(logn)
from bisect import bisect_left

def BST_Find_Index(sortedList,n):
  idx = bisect_left(sortedList,n)
  if idx < len(sortedList) and sortedList[idx] == n:
    return idx
  else:
    raise Exception('n is not in the list')

# Function to find all primes < n using "Sieve Method"
# Time complexity = O(sqrt(n))
def Sieve_All_Primes(n):
  sieve = [True] * n
  for i in range(3,int(n**0.5)+1,2):
    if sieve[i]:
      sieve[i*i::2*i]=[False]*((n-i*i-1)//(2*i)+1)
  
  # Output the list of all primes
  return [2] + [i for i in range(3,n,2) if sieve[i]]

from itertools import permutations
def Create_Perm_Num(n):
  perm = permutations(list(str(n)))
  permNum = set()
  for i in perm:
    permNum.add(int(''.join(i)))
  return permNum

def Prime_Permutations_K3(n):
  lim = 10**6
  primes = Sieve_All_Primes(lim)[167:]

  for i in primes:
    if i >= n:
      break

    permNum = sorted(Create_Perm_Num(i))

    for j in permNum[BST_Find_Index(permNum,i)+1:]:
      if BST_Find(primes,j) == True:
        k = (j-i) + j
        if BST_Find(primes,k)==True and BST_Find(permNum,k)==True:
          print(*[i,j,k], sep='')
          #print([i,j,k])

def Prime_Permutations_K4(n):
  lim = 10**6
  primes = Sieve_All_Primes(lim)[167:]

  for i in primes:
    if i >= n:
      break

    permNum = sorted(Create_Perm_Num(i))

    for j in permNum[BST_Find_Index(permNum,i)+1:]:
      if BST_Find(primes,j) == True:
        k = (j-i) + j
        if BST_Find(primes,k) == True and BST_Find(permNum,k) == True:
          l = (k-j) + k
          if BST_Find(primes,l) == True and BST_Find(permNum,l) == True:
            #print([i,j,k,l])
            print(*[i,j,k,l], sep='')
            
def Prime_Permutations(n,k):
  if k == 3:
    Prime_Permutations_K3(n)
  elif k == 4:
    Prime_Permutations_K4(n)

if __name__ =='__main__':
  n,k = map(int,input().split())
  Prime_Permutations(n,k)
```
